```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (8, 4)
plt.rcParams["axes.grid"] = True
```

```python
def load_membership_results(results_dir: str) -> pd.DataFrame:
    """
    Load all membership_mnist_dp*.json files from a directory and return a DataFrame.

    Expected JSON format per file:
        {
            "ks_stat": float,
            "ks_p_value": float,
            "welch_t_stat": float,
            "welch_t_p_value": float,
            "total_variation_distance": float,
            "empirical_epsilon": float
        }

    The filenames are expected to look like:
        membership_mnist_dpFalse.json
        membership_mnist_dpTrue.json
    """
    results_dir = Path(results_dir)
    json_files = sorted(results_dir.glob("membership_mnist_dp*.json"))

    rows = []
    for path in json_files:
        with path.open("r", encoding="utf-8") as f:
            data = json.load(f)

        name = path.name
        # Extract part after "dp" and before ".json", e.g. "True"/"False" or "true"/"false"
        dp_part = name.split("dp", maxsplit=1)[1].split(".json")[0]
        dp_flag = dp_part.lower() == "true"

        row = {
            "file": name,
            "dp": dp_flag,
            "ks_stat": data.get("ks_stat"),
            "ks_p_value": data.get("ks_p_value"),
            "welch_t_stat": data.get("welch_t_stat"),
            "welch_t_p_value": data.get("welch_t_p_value"),
            "total_variation_distance": data.get("total_variation_distance"),
            "empirical_epsilon": data.get("empirical_epsilon"),
        }
        rows.append(row)

    df = pd.DataFrame(rows)
    # Optional: sort so dpFalse appears first, then dpTrue
    if not df.empty:
        df = df.sort_values(by="dp").reset_index(drop=True)
    return df
```

```python
RESULTS_DIR = "results"  # change if your JSON files are elsewhere

membership_df = load_membership_results(RESULTS_DIR)

print("Membership results:")
display(membership_df)
```

```python
metrics = [
    ("ks_stat", "KS statistic"),
    ("ks_p_value", "KS p-value"),
    ("welch_t_stat", "Welch t-statistic"),
    ("welch_t_p_value", "Welch t p-value"),
    ("total_variation_distance", "Total variation distance"),
    ("empirical_epsilon", "Empirical epsilon"),
]

for key, label in metrics:
    if key not in membership_df.columns:
        continue

    fig, ax = plt.subplots()
    x = np.arange(len(membership_df))

    ax.bar(x, membership_df[key].values)
    ax.set_xticks(x)
    ax.set_xticklabels(membership_df["dp"])
    ax.set_ylabel(label)
    ax.set_title(f"{label} for membership experiment (MNIST)")
    plt.tight_layout()
    plt.show()
```

```python
print("=== Quick comparison DP vs no-DP ===")
for key, label in metrics:
    if key not in membership_df.columns:
        continue

    print(f"\n{label}:")
    for _, row in membership_df.iterrows():
        print(f"  {row['dp_label']:>5}: {row[key]:.4g}")
```
